```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(20,5))

plt.subplot(131)
plt.scatter(df['TV'],df['sales'])
plt.xlabel("Advertising on TV")
plt.ylabel("Sales")

plt.subplot(132)
plt.scatter(df['radio'],df['sales'])
plt.xlabel("Advertising on Radio")
plt.ylabel("Sales")

plt.subplot(133)
plt.scatter(df['newspaper'],df['sales'])
plt.xlabel("Advertising on Newpaper")
plt.ylabel("Sales")
```

```python
df.corr()
```

```python
x = df.iloc[:,:-1] # SLide the predictors and targte into varibales X and Y resp.
print(x)
```

```python
y = df.iloc[:,3]
print(y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)        #Training your Model
model
```

```python
model.coef_   #Finding the coefficient
```

```python
model.intercept_    #Finding the Intercept
```

```python
y_pred = model.predict(x)
print(y_pred[:5])
type(y_pred)
```

```python
result = pd.DataFrame({'actual outcome':y,'predicted values':y_pred})
result.head()
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y,y_pred)
```

```python
import numpy as np
x_test = np.array([[100,10,50]])
new = pd.DataFrame(x_test)
```

```python
y_pred_test = model.predict(new)
print(y_pred_test)
```
